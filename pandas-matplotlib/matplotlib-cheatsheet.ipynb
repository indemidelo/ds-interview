# `Matplotlib` cheatsheet

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

The basic steps to creating plots with mpl are:
1. Prepare data
2. Create plot
3. Plot
4. Customize plot
5. Save plot
6. Show plot

### 1. Prepare Data

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
```

### 2. Create plot

```python
fig = plt.figure()
```

### 3. Plot

```python
ax = fig.add_subplot(111)
```

### 4. Customize plot

```python
ax.plot(x, y, color='lightblue', linewidth=3)
ax.scatter([2, 4, 6],
          [5, 15, 25],
          color='darkgreen',
          marker='^')
ax.set_xlim(1, 6.5)
```

### 5. Save plot

```python
# plt.savefig('figure.png')
```

### 6. Show plot

```python
plt.show() # quando sono in uno script per vedere il grafico lancio questo
```

```python
fig # quando sono in un notebook per vedere il grafico lancio questo
```

## Multiple Axes

```python
fig = plt.figure()
fig2 = plt.figure(figsize=plt.figaspect(2.0))
```

All plotting is done with respect to an `Axes`. In most cases, a subplot will fit your needs. A subplot is an axes on a grid system.

```python
# fig.add_axes()
ax1 = fig.add_subplot(221) # row-col-num
ax3 = fig.add_subplot(122)
ax1.plot(x, y)
ax3.plot(x, y)
fig
```

```python
fig3, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y)
```

```python
fig4, axes = plt.subplots(nrows=2, ncols=2)
axes[1][1].plot(x, y)
```

## General recipe

If I need only one plot in a figure I go with classic `plot`:

```python
plt.figure()
plt.plot(x, y)
plt.show()
```

If I need multiple plots in a single figure I go with `subplots`:

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0][0].plot(x, y)
axes[0][1].plot(x, y, color='red', linewidth=4)
axes[1][0].plot(y, x)
axes[1][1].plot(x, np.array(y)^2 + 1)
axes[1][1].plot([1, 4], [10, 30], lw=2, linestyle='--') # I can draw intervals with their extremes
axes[1][0].plot([15, 25], [2, 3])
axes[0][0].scatter([2, 2.5, 3], [15, 15, 15]) # Scatter draws points
```
