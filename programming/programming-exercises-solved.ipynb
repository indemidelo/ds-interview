## Find all permutations of the string **s** in the string **b**

The idea is to check at each character of `b` if it is the first letter of a permutation of `s`. To see this a proxy version of `s` is created (`s_proxy`) and is shrinked until all the letters of `s` -- in any given order -- are found in `b`.

Otherwise, the search continues and the letters checked are skipped.

```python
from copy import deepcopy

def find_permutations(b, s):
    j_b = 0
    while j_b < len(b):
        cont = len(s)
        s_proxy = deepcopy(s)
        exit_cond = False
        while not exit_cond and cont > 0:
            letter = b[j_b]
            if letter in s_proxy:
                cont -= 1
                s_proxy = ''.join([c for c in s_proxy if c != letter])
                print(s_proxy, cont)
                j_b += 1
            else:
                exit_cond = True

        print(j_b, exit_cond)

        if exit_cond:
            j_b += len(s) - cont + 1
        else:
            return True
        
    return False



s = 'ciao'
b = 'dkeoiacjdedeodoe'
print(find_permutations(b, s))
```
